```python
import pandas as pd
import numpy as np
import scipy.stats as stats
```

```python
df = pd.read_csv('homepage-experiment-data.csv')
```

```python
df.head()
```

```python
df.shape
```

```python
# get number of trials and number of 'successes'
n_obs = df['Control Cookies'].sum() + df['Experiment Cookies'].sum()
n_control = df['Control Cookies'].sum()
```

Null Hypothesis: there is no difference between people downloading in both groups - p is 0.5 balanced

```python
# Compute a z-score and p-value
p = 0.5
sd = np.sqrt(p * (1-p) * n_obs)

z = ((n_control + 0.5) - p * n_obs) / sd

print(z)
print('P-value between groups - {}'.format(2 * stats.norm.cdf(z)))
```

```python
# number of observations in contrl group
n_cont = df['Control Cookies'].sum()
```

```python
#downloads in control group
n_cont_down = df['Control Downloads'].sum()
#proportion of purchases in control group
p_null_down = n_cont_down / n_cont
```

```python
# number of observations in experiment group
n_exp = df['Experiment Cookies'].sum()
```

```python
#downloads in experiment group
n_exp_down = df['Experiment Downloads'].sum()
#proportion of purchases in control group
p_alt_down = n_exp_down / n_exp
```

```python
print('rate for downloads control', p_null_down)
print('rate for downloads experiment', p_alt_down)
```

```python
p_overall = (n_exp_down + n_cont_down) / (n_exp + n_cont)
```

```python
# Compute a z-score and p-value
#standard error
sd = np.sqrt(p_overall * (1-p_overall)*(1/n_exp + 1/n_cont))

z = (p_alt_down - p_null_down) / sd

print(z)
print('P-value between groups - {}'.format(1-stats.norm.cdf(z)))
```

p-value and z-score for the evaluation metric: License Purchase Rate

```python
# number of observations in contrl group. 21 days
n_cont_21 = df[df['Day']<22]['Control Cookies'].sum()
```

```python
#purchase in control group
n_cont_lic = df['Control Licenses'].sum()
#proportion of purchases in control group
p_null_lic = n_cont_lic / n_cont_21
```

```python
# number of observations in experiment group. 21 days
n_exp_21 = df[df['Day']<22]['Experiment Cookies'].sum()
```

```python
#purchase in experiment group
n_exp_lic = df['Experiment Licenses'].sum()
#proportion of purchases in experiment group
p_alt_lic = n_exp_lic / n_exp_21
```

```python
print('P-value control group', p_null_lic)
print('P-value experiment group', p_alt_lic)
```

```python
p_alt_lic-p_null_lic
```

```python
p_total = (n_exp_lic + n_cont_lic) / (n_exp_21 + n_cont_21)
```

```python
print('P-value total', p_total)
```

```python
# Compute a z-score and p-value
#standard error
sd = np.sqrt(p_total * (1-p_total) * (1/n_cont_21 + 1/n_exp_21))

z = (p_alt_lic-p_null_lic) / sd

print(z)
print('P-value between groups - {}'.format(1- stats.norm.cdf(z)))
```
